```python
from sklearn.pipeline import Pipeline as p
from sklearn.preprocessing import StandardScaler,OneHotEncoder,OrdinalEncoder
from sklearn.compose import ColumnTransformer 

np=p([
    ('num',StandardScaler())
])
nl=p(
    [
        ('enc',OneHotEncoder())
    ]
)
nc=p(
    [
        ('lenc',OrdinalEncoder())
    ]
)
l=[]
l.extend(['Contract','PaymentMethod','MultipleLines'])
l

# num_col=num_col.drop(['Churn'])
num_col
cat_col=cat_col.drop(['customerID'])
cat_col
```

```python
prepro=ColumnTransformer([
    ('e',nl,l),('en',nc,cat_col),('s',np,num_col)
])
x = df.drop('Churn', axis=1)
y = df['Churn']
x = prepro.fit_transform(x)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
xt,xte,yt,yte=train_test_split(
    x,y,test_size=0.2,random_state=42
)

model=LogisticRegression(max_iter=100000)
model.fit(xt,yt)
yp=model.predict(xte)

from sklearn.metrics import accuracy_score
acc=accuracy_score(yte,yp)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(yte, yp)

print(cm)
acc*100
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(xte)

print(classification_report(yte, yp))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# ---------------- Decision Tree ----------------

dt = DecisionTreeClassifier(random_state=42)

dt.fit(xt, yt)

dt_pred = dt.predict(xte)

print("Decision Tree Accuracy:", accuracy_score(yte, dt_pred))
print("Decision Tree Confusion Matrix:")
print(confusion_matrix(yte, dt_pred))
print("Decision Tree Classification Report:")
print(classification_report(yte, dt_pred))


# ---------------- Random Forest ----------------

rf = RandomForestClassifier(random_state=42)

rf.fit(xt, yt)

# rf_pred = rf.predict(xte)
# print(rf.featureImportanc_)
importance = dt.feature_importances_

print(importance)

# print("Random Forest Accuracy:", accuracy_score(yte, rf_pred))
# print("Random Forest Confusion Matrix:")
# print(confusion_matrix(yte, rf_pred))
# print("Random Forest Classification Report:")
# print(classification_report(yte, rf_pred))
```

```python
df=pd.read_csv("Churn_Prediction.csv")
df=df.drop(['gender','Partner','MultipleLines','InternetService','StreamingTV','StreamingMovies','PaymentMethod','TotalCharges'],axis=1)
df.info()
```

```python
df.head()
df=df.drop('customerID',axis=1)
df.head()
from sklearn.preprocessing import OrdinalEncoder
d=OrdinalEncoder()
d.fit(df[['Contract','DeviceProtection','TechSupport','OnlineBackup','OnlineSecurity']])
df[['Contract','DeviceProtection','TechSupport','OnlineBackup','OnlineSecurity']]=d.transform(df[['Contract','DeviceProtection','TechSupport','OnlineBackup','OnlineSecurity']])
```

```python
num_col=df.select_dtypes(include=['number']).columns
num_col
df['Churn']=df['Churn'].map({'Yes':1,'No':0})


```

```python
df.head()
df=pd.get_dummies(df,dtype=int)
df.head()
```

```python
x = df.drop('Churn', axis=1)
y = df['Churn']
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
xt,xte,yt,yte=train_test_split(
    x,y,test_size=0.2,random_state=42
)
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42,class_weight={0:1, 1:3}),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42)
}

for name, model in models.items():

    model.fit(xt, yt)

    y_pred = model.predict(xte)

    print("="*50)
    print(name)
    print("="*50)
    print("Accuracy:", accuracy_score(yte, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(yte, y_pred))
    print("Classification Report:")
    print(classification_report(yte, y_pred))
```

```python
cat_cols = ["gender", "Partner", "Dependents","Contract"]

feature_scores = {}

for col in cat_cols:
    churn_rate = pd.crosstab(
        df[col],
        df["Churn"],
        normalize="index"
    ) * 100

    difference = churn_rate["Yes"].max() - churn_rate["Yes"].min()

    feature_scores[col] = difference

scores = (
    pd.Series(feature_scores)
      .sort_values(ascending=False)
)

print(scores)
df.info()
```
